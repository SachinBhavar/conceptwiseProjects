```python
#Bar graph
import matplotlib.pyplot as plt
plt.bar([0.24,0.80,0.70,0.90,0.50,0.40],[20,70,50,80,10,90],label='population',color='r',width=0.05)
plt.xlabel('population')
plt.ylabel('frequency')
plt.show()
```

```python
#histogram
#arrays ,long list etc, falls in some sort of distribution 
age=[12,13,23,34,45,56,67,78,98]
bins=[0,10,20,30,100]
```

```python
plt.hist(age,bins=bins,histtype='bar',rwidth=0.1)
```

```python
#scatter plot , to show relationship 
weights=np.array([1,2,3,4,5,6,7,8])
speed=np.array([60,40,30,25,20,15,12,9])
age=np.array([22,33,44,55,23,12,45,11])
```

```python
plt.scatter(weights,speed,marker='*')
plt.show()
```

```python
plt.plot([],[],color='g',label=weights,linewidth=5)
plt.plot([],[],color='b',label=speed,linewidth=5)
plt.plot([],[],color='r',label=age,linewidth=5)

plt.stackplot(weights,speed,age)
plt.show()
#stackplot is used to plot multiple variable under merging condition
```
