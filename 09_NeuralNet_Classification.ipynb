```python
from sklearn.model_selection import train_test_split
```

```python
x = df_spine.drop(['class'],axis=1)
y = df_spine['class']
```

```python
x
```

```python
y
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2)
```

```python
x_train.shape, y_train.shape, x_test.shape, y_test.shape
```

```python
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import roc_auc_score
from sklearn.model_selection import RandomizedSearchCV
```

```python
parameters={
'learning_rate': ["constant", "invscaling", "adaptive"],
'hidden_layer_sizes': [(5,10,5),(20,10),(10,20),(50,50),(100,100)],
'alpha': [0.3,.1,.01,0.001,0.0001],
'activation': ["logistic", "relu", "tanh"],
    'solver':['lbfgs']
}
clf=MLPClassifier()
random_search=RandomizedSearchCV(clf,n_iter=5,cv=10,param_distributions=parameters,
                                 scoring='roc_auc',random_state=2,n_jobs=-1,verbose=20)
```

```python
random_search.fit(x_train, y_train)
```

```python
random_search.best_estimator_
```

```python
def report(results, n_top=3):
    for i in range(1, n_top + 1):
        candidates = np.flatnonzero(results['rank_test_score'] == i)
        for candidate in candidates:
            print("Model with rank: {0}".format(i))
            print("Mean validation score: {0:.3f} (std: {1:.5f})".format(
                  results['mean_test_score'][candidate],
                  results['std_test_score'][candidate]))
            print("Parameters: {0}".format(results['params'][candidate]))
            print("")
```

```python
import numpy as np
report(random_search.cv_results_,5)
```

```python
mlp=random_search.best_estimator_
mlp
```

```python
print("Training accuracy",mlp.score(x_train, y_train))
```

```python
mlp.fit(x_train,y_train)
```

```python
y_pred = mlp.predict(x_test)
```

```python
pred_results = pd.DataFrame({'actual':y_test, 'pred':y_pred})
```

```python
pred_results.sample(10)
```

```python
from sklearn.metrics import accuracy_score
```

```python
print("Testing accuracy",accuracy_score(y_test,y_pred))
```
